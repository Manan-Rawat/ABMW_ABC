```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from itertools import product
import seaborn as sns

np.random.seed(42)

n_samples = 20
n_features = 8
beta_true = np.array([3, 1.5, 0, 0, 2, 0, 0, 0])
sigma = 3
rho = 0.5

def generate_correlated_data(n, p, rho):
    cov = rho ** np.abs(np.subtract.outer(np.arange(p), np.arange(p)))
    return np.random.multivariate_normal(mean=np.zeros(p), cov=cov, size=n)

datasets = []
for _ in range(50):
    X = generate_correlated_data(n_samples, n_features, rho)
    noise = sigma * np.random.randn(n_samples)
    y = X @ beta_true + noise
    datasets.append((X, y))

model_space = list(product([0, 1], repeat=n_features))

def calibrate_epsilon(y_obs, X_obs, model_space, n_samples=10000, target_acceptance=0.0005):
    eps_grid = np.linspace(1.0, 10.0, 50)
    acceptance_rates = []
    for eps in eps_grid:
        accepted = 0
        for _ in range(n_samples):
            m = model_space[np.random.randint(len(model_space))]
            active_idx = np.where(np.array(m) == 1)[0]
            if len(active_idx) == 0:
                continue
            theta = np.random.uniform(-10, 10, len(active_idx))
            y_sim = X_obs[:, active_idx] @ theta
            dist = np.sqrt(mean_squared_error(y_obs, y_sim))
            if dist < eps:
                accepted += 1
        acceptance_rates.append(accepted / n_samples)
    idx = np.argmin(np.abs(np.array(acceptance_rates) - target_acceptance))
    best_eps = eps_grid[idx]
    plt.plot(eps_grid, acceptance_rates, marker='o')
    plt.axvline(best_eps, color='r', linestyle='--')
    plt.title("Epsilon Calibration")
    plt.xlabel("Epsilon")
    plt.ylabel("Acceptance Rate")
    plt.grid()
    plt.show()
    print(f"Calibrated epsilon: {best_eps:.4f}")
    return best_eps

X0, y0 = datasets[0]
epsilon = calibrate_epsilon(y0, X0, model_space)

def abc_rjmcmc(X, y, model_space, epsilon, n_iter=10000, model_probs=None):
    model_stats = {i: {"count": 0, "params": [], "rmse": []} for i in range(len(model_space))}
    for _ in range(n_iter):
        if model_probs is None:
            m_idx = np.random.randint(len(model_space))
        else:
            m_idx = np.random.choice(len(model_space), p=model_probs)
        m = model_space[m_idx]
        active_idx = np.where(np.array(m) == 1)[0]
        if len(active_idx) == 0:
            continue
        theta = np.random.uniform(-10, 10, len(active_idx))
        y_sim = X[:, active_idx] @ theta
        dist = np.sqrt(mean_squared_error(y, y_sim))
        if dist < epsilon:
            model_stats[m_idx]["count"] += 1
            model_stats[m_idx]["params"].append(theta)
            model_stats[m_idx]["rmse"].append(dist)
    return model_stats

def calculate_weights(perf_dict, temp=0.5):
    scores = {}
    for i, perf in perf_dict.items():
        if perf["rmse"]:
            scores[i] = np.exp(-np.mean(perf["rmse"]) / temp)
        else:
            scores[i] = 0
    total = sum(scores.values())
    if total == 0:
        return np.ones(len(perf_dict)) / len(perf_dict)
    return np.array([scores[i] / total for i in range(len(perf_dict))])

print("Running initial ABC-RJMCMC for model probabilities...")
initial_stats = abc_rjmcmc(X0, y0, model_space, epsilon, n_iter=10000)
model_probs = calculate_weights(initial_stats, temp=0.5)

all_results = []
for i, (X, y) in enumerate(datasets):
    print(f"Running ABC-RJMCMC on dataset {i+1}/50")
    stats = abc_rjmcmc(X, y, model_space, epsilon, model_probs=model_probs)
    all_results.append(stats)
    model_probs = calculate_weights(stats, temp=0.5)

model_counts = np.zeros(len(model_space))
for result in all_results:
    for i, stat in result.items():
        model_counts[i] += stat["count"]

model_probs = model_counts / np.sum(model_counts)
top_models = sorted(
    [(i, model_space[i], prob) for i, prob in enumerate(model_probs)],
    key=lambda x: x[2],
    reverse=True
)

print("\nTop 10 models by posterior probability:")
for rank, (i, m, p) in enumerate(top_models[:10], 1):
    print(f"{rank:2d}: Model {i:03d} | Active Vars: {m} | Posterior: {p:.4f}")

plt.figure(figsize=(12, 5))
plt.bar(range(10), [p for _, _, p in top_models[:10]], tick_label=[f"M{i}" for i, _, _ in top_models[:10]])
plt.ylabel("Posterior Probability")
plt.title("Top 10 Models")
plt.grid(True)
plt.show()

# --- Identify the True Model and its Index ---
true_model_spec = tuple((beta_true != 0).astype(int))
true_model_index = model_space.index(true_model_spec)
print(f"\nTrue Model:")
print(f"  Specification: {true_model_spec}")
print(f"  Index: {true_model_index}")

def plot_top_model_posteriors(top_models, all_results, model_space, beta_true, n_top=10):
    for rank, (model_idx, model_spec, _) in enumerate(top_models[:n_top], 1):
        active_idx = np.where(np.array(model_spec) == 1)[0]
        param_samples = []
        for result in all_results:
            param_samples.extend(result[model_idx]["params"])

        if not param_samples:
            print(f"No accepted parameters for Model {model_idx}")
            continue

        param_samples = np.array(param_samples)
        n_params = param_samples.shape[1]

        plt.figure(figsize=(4 * n_params, 4))
        print(f"\n--- Model {rank} (Index: {model_idx}, Specification: {model_spec}) ---")
        for i in range(n_params):
            plt.subplot(1, n_params, i + 1)
            try:
                sns.kdeplot(param_samples[:, i], fill=True, color="skyblue", linewidth=2)
            except:
                plt.hist(param_samples[:, i], bins=30, density=True, alpha=0.7, color='skyblue')

            # Get the true parameter value for this active feature
            true_param_value = beta_true[active_idx[i]]
            plt.axvline(true_param_value, color='red', linestyle='--', label=f'True β{active_idx[i]+1}={true_param_value:.2f}')

            # Calculate and print posterior mean and median
            posterior_mean = np.mean(param_samples[:, i])
            posterior_median = np.median(param_samples[:, i])

            print(f"  Parameter β{active_idx[i]+1} (True: {true_param_value:.2f}):")
            print(f"    Posterior Mean: {posterior_mean:.4f}")
            print(f"    Posterior Median: {posterior_median:.4f}")
            print(f"    Difference (Mean - True): {posterior_mean - true_param_value:.4f}")
            print(f"    Difference (Median - True): {posterior_median - true_param_value:.4f}")

            plt.title(f"Model {model_idx} | β{active_idx[i]+1}")
            plt.xlabel(f"β{active_idx[i]+1}")
            plt.grid(True)
            plt.legend() # Show legend for the true value line

        plt.suptitle(f"Posterior Densities for Top Model {rank} (Model {model_idx})", fontsize=14)
        plt.tight_layout(rect=[0, 0, 1, 0.93])
        plt.show()

# Call the modified plotting function
plot_top_model_posteriors(top_models, all_results, model_space, beta_true)
```
